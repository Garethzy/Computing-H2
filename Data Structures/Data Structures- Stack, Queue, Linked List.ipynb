```python
#Longer node class with getter and setter methods

class Node:
    def __init__(self, data, next=None):
        self._data = data
        self._next = next
        
    def getData(self):
        return self._data
    
    def setData(self, data):
        self._data = data
        
    def getNext(self):
        return self._next
    
    def setNext(self, node):
        self._next = node
        
    def __str__(self):
        ret = f"Data: {self.getData()}"
        if self.getNext():
            ret += f", Next: {self.getNext().getData()}"
        else:
            ret += ", Next: None"
        return ret
    
ptr = None
for count in range(5):
    ptr = Node(count+1, ptr)
    
while ptr != None:
    print(ptr)
    ptr = ptr.getNext()
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
    def create(self):
        for i in range(4):
            self.head = Node(str(i+1), self.head)
    
    def __str__(self):
        probe = self.head
        ret = ''
        while probe != None:
            ret += probe.__str__() + '\n'
            probe = probe.getNext()
        return ret
    
    def insertStart(self, val):
        self.head = Node(val, self.head)
        
    def insertEnd(self, val):
        probe = self.head
        while probe.getNext() != None:
            probe = probe.getNext()
        probe.setNext(Node(val))
    
    def insertAny(self, val, index):
        probe = self.head
        if index == 0:
            self.head = Node(val, self.head)
        else:
            for i in range(index - 1):
                probe = probe.getNext()
            probe.setNext(Node(val, probe.getNext()))
            
    def removeAny(self, index):
        probe = self.head
        if index == 0:
            self.head = probe.getNext()
        else:
            for i in range(index - 1):
                probe = probe.getNext()
            probe.setNext(probe.getNext().getNext())
        

x = LinkedList()
x.create()
print(x)
x.insertEnd(102)
print(x)
x.removeAny(3)
print(x)
x.insertAny(21, 2)
print(x)
```

```python
#Taking the first index in the list to be 0
#Assuming value in all operations refer to data and not nodes

class LinkedList:
    def __init__(self):
        self.head = None
        
    def create(self):
        for i in range(4):
            self.head = Node(str(i + 1), self.head)
            
    def __str__(self):
        probe = self.head
        ret = ''
        while probe != None:
            ret += probe.__str__() + '\n'
            probe = probe.next
        return(ret)

    def insertStart(self, val):
        self.head = Node(val, self.head)

    def insertEnd(self, val):
        probe = self.head
        while probe.next != None:
            probe = probe.next
        probe.next = Node(val, None)

    def insertAny(self, val, index):
        probe = self.head
        if index == 0:
            self.head = Node(val, self.head)
        else:
            for i in range(index - 1):
                probe = probe.next
            probe.next = Node(val, probe.next)

    def find(self, val):
        probe = self.head
        pos = 0
        while probe != None:
            if probe.data == val:
                return(f"Found, index: {pos}")
            else:
                probe = probe.next
                pos += 1
        return("Not found")

    #Insert val before the last index
    def insertbefore(self, val):
        newNode = Node(val, None)
        probe = self.head
        while probe.next.next != None:
            probe = probe.next
        last = probe.next
        probe.next = newNode
        newNode.next = last
        

    def removeStart(self):
        self.head = self.head.next

    def removeEnd(self):
        probe = self.head
        while probe.next.next != None:
            probe = probe.next
        probe.next = None

    def removeAny(self, index):
        if index == 0:
            self.head = self.head.next
        else:
            probe = self.head
            for i in range(index-1):
                probe = probe.next
            #at this point, probe points to the Node before index
            probe.next = probe.next.next
```

# Stack ADT

LIFO (Last In First Out structure) 
Operations are restricted to one end (called the top).
Operations:
- push: Inserts item to the top
- pop: Removes item at the top
- peek: Returns item at the top
- isEmpty
- __len__
- __str__

Implemented using either
- Array (Using python List)
      Built around an array called items and two integers called top and           size. 
      Default value of top is -1 and size is 0.
      To push an item, increment top and size. Store item at items[top]
- Linked Structure (Linked List made of Nodes)

Applications: Infix and Postfix conversion and evaluation

https://drive.google.com/file/d/12Xc1scfJcd9yMB3bcr1H3KwmlRKK3At3/view

```python
#Array Stack Implementation

class ArrayStack():
    CAPACITY = 12
    
    def __init__(self):
        self.items = [''] * ArrayStack.CAPACITY
        self.top = -1
        self.size = 0
        
    def push(self, item):
        if self.size == ArrayStack.CAPACITY:
            return("Stack is full")
        else:
            self.top += 1
            self.size += 1
            self.items[self.top] = item
            
    def pop(self):
        if self.isEmpty():
            return "Stack is Empty."
        else:
            removed = self.items[self.top]
            self.top -= 1
            self.size -= 1
            return f"Removed {removed}"
        
    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        else:
            return self.items[self.top]
        
    def isEmpty(self):
        return (self.size == 0)
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        ret = ''
        for i in range(self.size):
            ret += str(self.items[i]) + " "
        ret += 'TOP'
        return ret
    
    
x = ArrayStack()
x.push('1')
x.push('2')
x.push('3')
x.push('4')

print(x)
print('\nPOP:')
print(x)
print(x.pop())
print('\nPEEK:')
print(x)
print(x.peek())
print('\nFINAL:')
print(x)
```

```python
#Official linked list implementation with getter and setter methods

class LinkedListStack():
    def __init__(self):
        self._top = None
        self._size = 0
        
    def getTop(self):
        return self._top
    
    def setTop(self, node):
        self._top = node
    
    def getSize(self):
        return self._size
    
    def setSize(self, node):
        self._size = node
        
    def _add(self):
        self._size += 1
    
    def _remove(self):
        self._size -= 1
        
    def push(self, item):
        if self.getTop():
            newNode = Node(item, self.getTop())
            self.setTop(newNode)
        else:
            self.setTop(Node(item))
        self._add()
        
    def pop(self):
        if self.getTop():
            data = self.getTop().getData()
            self._remove()
            if self.getTop().getNext():
                self.setTop(self.getTop().getNext())
            else:
                self.setTop(None)
            return data
        else:
            return None
        
    def isEmpty(self):
        if self.getSize() == 0:
            return True
        else:
            return False
        
    def peek(self):
        if self.getTop():
            return self.getTop().getData()
        else:
            return None
        
    def size(self):
        return self.getSize()
    
    def __str__(self):
        ret = 'TOP '
        if self.getTop():
            probe = self.getTop()
            while probe:
                ret += probe.getData() + " "
                probe = probe.getNext()
        else:
            ret = "None"
        return ret
    

if __name__ == "__main__":
    stk = LinkedListStack()
    print("PUSH")
    stk.push('1')
    stk.push('2')
    stk.push('3')
    print(stk)
    print("\nPOP")
    stk.pop()
    print(stk)
    print("\nPEEK")
    print(stk.peek())
```

```python
#Shorter linked list implementation

class LinkedStack():
    def __init__(self):
        self.top = None    #Similar to head of a linked list
        self.size = 0
        
    def push(self, item):
        self.top = Node(item, self.top)
        self.size += 1
    
    def pop(self):
        if self.isEmpty():
            return "Is Empty"
        else:
            removed = self.top.data
            self.top = self.top.next
            self.size -= 1
            return f"Removed {removed}"
        
    def peek(self):
        if self.isEmpty():
            return "Is Empty"
        else:
            return self.top.data
        
    def isEmpty(self):
        return self.size == 0
            
    def __len__(self):
        return self.size
        
    def __str__(self):
        probe = self.top
        ret =  'TOP| '
        while probe != None:
            ret += f"{probe.data} "
            probe = probe.next
        return(ret + '|')
```

# Queue

FIFO(First-In First-Out) protocol
Operations:
- Enqueue: Adds an item to the rear. 
- Dequeue: Removal of an item from the front.
- Peek: Returns item at the front of queue
- isEmpty
- __len__
- __str__

Implementations:
- ArrayQueue 
      Uses Python List
      Using a Circular Array Implementation: 
          rear starts at -1; front starts at 0; size at 0
          front chases rear pointer through the array
          When a pointer is about to run off the end, it is reset to 0
          This wraps the queue around to the beginning to the array.
- LinkedQueue (Linear)
      Using Nodes
      Create front and rear with initial value None and size to 0.

```python
class ArrayQueue:
    CAPACITY = 9
    def __init__(self):
        self.size = 0
        self.items = [''] * self.CAPACITY
        self.rear = -1
        self.front = 0
        
    def enqueue(self, item):
        if self.size == self.CAPACITY:
            return 'Queue is full'
        else:
            if self.rear == self.CAPACITY - 1:
                self.rear = 0
            else:
                self.rear += 1
            self.items[self.rear] = item
            self.size += 1
    
    def dequeue(self):
        if self.size == 0:
            return 'Empty'
        else:
            removed = self.items[self.front]
            if self.front == self.CAPACITY - 1:
                self.front = 0
            else:
                self.front += 1
            self.size -= 1
            return removed
        
    def isEmpty(self):
        return (self.size == 0)
        
    def peek(self):
        if self.isEmpty():
            return -1
        else:
            return self.items[self.front]
    
    def __str__(self):
        ret = 'FRONT| '
        if self.front > self.rear:
            for i in range(self.front, self.CAPACITY):
                ret += str(self.items[i]) + ' '
            for i in range(self.rear + 1):
                ret += str(self.items[i]) + ' '
        else:
            for i in range(self.front, self.rear + 1):
                ret += str(self.items[i]) + ' '
        return ret + '|REAR'
                
x = ArrayQueue()
x.enqueue('1')
x.enqueue('2')
x.enqueue('3')
x.enqueue('4')
x.enqueue('5')
x.enqueue('6')
x.enqueue('7')
x.enqueue('8')

print('QUEUE')
print(x.items)
print(x)

print('\nDEQUEUE')
print("Removed:", x.dequeue())
print("Removed:", x.dequeue())
print("Removed:", x.dequeue())
print(x.items)
print(x)

print('\nENQUEUE')
x.enqueue('9')
x.enqueue('10')
# x.enqueue('11')
# x.enqueue('12')
# # When trying to insert 13, there is an error since queue is full
# print(x.enqueue('13'))    
print(x.items)
print(x)

print('\nPEEK')
print("Top element: ", x.peek())
print(x.items)
print(x)
```

```python
#Linked List Queue Implementation

class LinkedQueue():
    def __init__(self):
        self.front = None
        self.rear = None
        self.size = 0
        
    def enqueue(self, item):
        newNode = Node(item, None)
        if self.isEmpty():
            self.front = newNode
        else:
            self.rear.next = newNode
        self.rear = newNode
        self.size += 1
        
    def dequeue(self):
        if self.isEmpty():
            return "Queue is Empty."
        else:
            removed = self.front.data
            self.front = self.front.next
            if self.front is None:
                self.rear = None
            self.size -= 1
            return removed
            
    def peek(self):
        if self.isEmpty():
            return "Queue is Empty"
        else:
            return self.front.data
        
    def isEmpty(self):
        return (self.size == 0)
    
    def __len__(self):
        return (self.size)
    
    def __str__(self):
        ret = 'Front| '
        probe = self.front
        while probe != None:
            ret += str(probe.data) + " "
            probe = probe.next
        return ret + '|Rear'
```

```python
x = LinkedQueue()
x.enqueue('1')
x.enqueue('2')
x.enqueue('3')

print('FULL')
print(x)

print('\nDEQUEUE')
print("Removed:", x.dequeue())
print(x)

print('\nUPDATED')
print(x)
```
