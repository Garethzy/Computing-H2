### 1. Bubble Sort method
Steps:
- For each pass, compare adjacent elements and exchange their values if the first value is greater than the second. At the end of each pass, the largest element is at the end of the current sublist.
- After first pass, the end of the list(A[n-1]) is sorted while front remains unordered
- Uses a lastExchangeIndex to determine when to end, end when LEI=0. Elements in the tail of the list from A [lastExchangeIndex + 1] to A [n-1] are sorted.

 Comments:
- Requires up to (n-1) passes for array A with n elements

Time Complexity: O(n^2)

````python
def swap(A, j, i):
    temp = A[j]
    A[j] = A[i]
    A[i] = temp
    

def BubbleSort(A, n):
    i = len(A) - 1 #index of the last element of the unsorted sublist
    
    #if i = 0 it means that all the elements have been sorted 
    #since the last  unsorted element is the only unsorted element
    
    while i > 0:  
        lastExchangeIndex = 0
        for j in range(i): #looping through the unsorted sublist
            if A[j] > A[j + 1]:
                swap(A, j, j + 1)
                lastExchangeIndex = j
                
        i = lastExchangeIndex   #index of last element in unsorted sublist
    print(A)
    
BubbleSort([50, 20, 40, 75, 35], 5)
````

### 2. Insertion Sort
- In place comparison based algorithm
- A sublist is maintained that is always sorted

The lower part is maintained to be sorted. The element which is to be inserted in this sorted sublist has to find the appropriate place and then has to be inserted there. 

The array is searched sequentially and unsorted items are moved and inserted into the sorted sublist.

Not suitable for large datasets.

Time Complexity: O(n^2)

````python
##Suggested version

def InsertionSort(A, n):
    for i in range(1, n):
        #index j scans down list from A[i-1]
        #to look for correct position to insert target
        target = A[i] 
        j = i
        
        #Locate insertion point by scanning downward
        #as long as target < A[j-1] and 
        #we have not encountered the beginning of the list
        while j>0 and target < A[j - 1]:
            #shifts each element up by one position
            A[j] = A[j-1]
            j -= 1
        #location found and target is inserted
        A[j] = target
        print(A)
InsertionSort([50, 20, 40, 75, 35], 5)  
````

### 3. Quick Sort

Fastest known sorting algorithm. It uses divide and conquer to gain the same advantages as merge sort while not using additional space.

- Step 1: Select a pivot value
- Step 2: Partition process- Find split point and move the items to the appropriate side of the list

Time complexity O(nlog2n)

````python
# Code with comments 

def split(A, low, high):
    mid = (low + high) // 2
    pivot = A[mid]
    print(f"pivot: {pivot}")

    swap(A, mid, low)
    left = low + 1
    right = high
    
    while left <= right:
        while left <= right and A[left] <= pivot:
            left += 1
        while A[right] > pivot:
            right -= 1
        if left < right:
            swap(A, left, right)
            print(A)
    print("left: ", left)
    print("right: ", right)
    print("1. ", A)
    #swaping pivot to the correct position at the final right position
    swap(A, low, right)    
    return right  #returning final position of pivot

def quickSort(A, low, high):
    if low < high: #list has more than one element
        #split into two sublists
        #pos is the final position of the pivot
        pos = split(A, low, high)
        print("2. ", A)
        print("LEFT")
        quickSort(A, low, pos-1) #quickSort left sublist
        print("RIGHT")
        quickSort(A, pos + 1, high) #quicksort right sublist
        return A
        
        
def swap(A, i, j):
    temp = A[i]
    A[i] = A[j]
    A[j] = temp
    
    
print(quickSort([800, 150, 300, 600, 550, 650, 400, 350, 450, 700], 0, 9))
````

### 4. Merge Sort

Divide and Conquer algorithm based on the idea of breaking down a list into several sub-lists until each sublist consists of a single element and merging those sublists in a manner that results into a sorted list.

Steps:
1. Divide the unsorted list into n sublists, each containing 1 element (a list of 1 element is considered sorted)
2. Take adjacent pairs of two singleton lists and merge them to form a new sorted sublist
3. Repeat the process till a single sorted list is obtained

Time complexity O(nlog2n)

````python
def MergeSort(values):
    if len(values) > 1:
        m = len(values) // 2
        left = values[:m]
        right = values[m:]
        left = MergeSort(left)
        right = MergeSort(right)
        values = []
        while len(left) > 0 and len(right) > 0:
            if left[0] < right[0]:
                values.append(left[0])
                left.pop(0)
            else:
                values.append(right[0])
                right.pop(0)

        for i in left:
            values.append(i)
        for i in right:
            values.append(i)
    return values

print(MergeSort([12, 11, 13, 5, 6, 7]))
````
